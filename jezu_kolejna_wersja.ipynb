#Real Time Detection


```python
#nie dzia≈Ça 
cap = cv2.VideoCapture(2)
while cap.isOpened():
  ret, frame = cap.read()

  #make detections
  results = model (frame)

  cv2.imshow('YOLO', np.squeeze(results.render()))
  #exit
  if cv2.waitKey(10) & 0xFF == ord('q'):
    cap.release()
    cv2.destroyAllWindows()
```
